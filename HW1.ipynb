```python
import json
from elasticsearch import Elasticsearch

es = Elasticsearch([{'host':'localhost','port':9200}])
print json.dumps(es.info(),indent=2)
```

```python
response = es.search(
    index="twitter2",
    body={
            "query": {
                "term": {
                    "text": "exploit"
                }
            },
            "from": 0,
            "size": 100
        }
)
print json.dumps(response,indent=2)
```

```python
#response =  json.dumps(response,indent=2)
for tweet in response['hits']['hits']:
   print tweet['_id']
  
```

```python
response = es.search(
    index="logstash*",
    body={
          "query": {
              "range": {
                  "@timestamp": {
                      "gt": "2016-10-01T15:39:11.000Z",
                      "lt": "2016-10-20T15:39:13.000Z"
                  }
              }
          },
          "from": 0,
          "size": 2
    }
)
print json.dumps(response,indent=2)
```

```python
response = es.search(
    index="logstash*",
    body={
          "query": {
            "bool": {
              "must": [
                {
                  "range": {
                    "@timestamp": {
                      "gt": "2016-10-02T15:39:11.000Z",
                      "lt": "2016-10-20T16:39:13.000Z"
                    }
                  }
                },
                {
                  "term": {
                    "message": "ppap"
                  }
                }
              ],
              "must_not": [],
              "should": []
            }
          },
          "from": 0,
          "size": 10
        }
)
for tweet in response['hits']['hits']:
   print tweet['_source']['user']
   print tweet['_source']['message']
   print tweet['_source']['@timestamp']
   print "\n"
```

```python
response = es.search(
    index="twitter2",
    body={
          "query": {
            "term": {
              "text": "exploit"
            }
          },
        "size":0,
        
        "aggs": {
            "user": {
                "terms": {
                    "field":"tweet.uid" ,
#                 "size": 5
              }
            }
          }
        }
)

# print response
print json.dumps(response,indent=2)


# for user in response['aggregations']['user']['buckets']:
#   print str(user['doc_count']) + "\t" + user['key'] 
```
